# Compras efetuadas com cartões de pagamento do Governo Federal

```python
#importando bibliotecas
import pandas as pd
import glob
```

```python
#caminho da pasta com todos os cvs
files = glob.glob(r'MEU CAMINHO ATÉ A PASTA')
```

```python
# pegando o primeiro csv como meumodelo
df = pd.read_csv(r'.\gastos_cartao\202003_CPGF.csv', encoding='Latin5',sep='\t')
```

```python
#unindo os csvs
df = pd.concat([pd.read_csv(file, encoding='Latin5', sep=';', header=None, skiprows=1) for file in files])
```

```python
df.head()
```

```python
#colocando cabeçalhos
df.columns = ['Código Órgão Superior','Nome Órgão Superior','Código Órgão Subordinado','Nome Órgão Subordinado','Código Unidade Gestora','Nome Unidade Gestora','Ano Extrato','Mês Extrato','CPF Portador','Nome Portador','CNPJ ou CPF Favorecido','Nome Favorecido','Transação','Data Transação','Valor Transação']
```

```python
#substituindo ',' por '.' e dizendo que a coluna 'Valor Transação' é numérica
df['Valor Transação'] = df['Valor Transação'].str.replace(',','.')
df['Valor Transação'] = df['Valor Transação'].astype(float)
```

```python
#transformando a coluna Data Transação para formato de data
df['Data Transação'] = pd.to_datetime(df['Data Transação'], dayfirst=True,format='%d/%m/%Y',errors='coerce')
```

```python
#selecionando apenas a partir de 11 de março de 2020
df = df[df['Data Transação'] > '11/03/2020']
```

```python
df
```

### Recortando para tema churrasco

```python
#criando df só com churrascarias, pesquisando a partir do termo 'churrasc'
churrasco = df[df['Nome Favorecido'].str.contains('CHURRASC')]
```

```python
#total de compras realizadas no período
churrasco.shape
```

```python
#qual foi o restaurante que mais faturou no período
churrasco.groupby('Nome Favorecido')['Valor Transação'].sum().sort_values(ascending=False)
```

```python
#qual foi o órgão superior que mais gastou
churrasco.groupby('Nome Órgão Superior')['Valor Transação'].sum().sort_values(ascending=False)
```

```python
#qual foi o portador de cartão corporativo que mais gastou na pandemia
churrasco.groupby('Nome Portador')['Valor Transação'].sum().sort_values(ascending=False)
```

```python
#quantos estabelecimentos foram visitados por portadores do cartão
lista_unica_estabelecimentos = churrasco['Nome Favorecido'].unique()
```

```python
#quantos estabelecimentos foram visitados por portadores do cartão
lista_unica_estabelecimentos.shape
```

```python
#exportando o csv completo para analisar melhor
churrasco.to_csv('churrasco_gastos_cartao_pandemia.csv',index=False)
```
